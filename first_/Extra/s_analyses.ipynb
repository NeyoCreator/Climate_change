```python
import numpy as np # linear algebra
import pandas as pd # data processing

##Data Cleaning
import nltk #
import string #to remove all punctuation marks in the corpus/ document
from nltk.tokenize import word_tokenize, TreebankWordTokenizer
from nltk import SnowballStemmer, PorterStemmer, LancasterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import re

from sklearn.model_selection import train_test_split

ps = nltk.PorterStemmer()

from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report, accuracy_score
# set plot style
sns.set()
```

# Loading Data 

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv('test.csv')
```

```python
#Show the full value of the dataset
pd.set_option("display.max_colwidth", -1)
```

# 0.remove urls

```python
### Remove urls and replace them with the word 'url-web'
#Our data doesn't seem to have any urls, adding code just in case
pattern_url = r'http[s]?://(?:[A-Za-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9A-Fa-f][0-9A-Fa-f]))+'
subs_url = r'url-web'
df_train['message'] = df_train['message'].replace(to_replace = pattern_url, value = subs_url, regex = True)

##Convert all lowercase to uppercase
df_train['message'] = df_train['message'].str.lower()

### Remove special characters, symbols
def remove_punctuation(post):
    return ''.join([l for l in post if l not in string.punctuation])

df_train['message'] = df_train['message'].apply(remove_punctuation)

df_train.head()
```

```python
df_train.head()
```

# 1. Punctuation

```python
import string 
string.punctuation
```

```python
#1.1 remove the punctuation
def remove_punctuation(text):
    no_punct=[words for words in text if words not in string.punctuation]
    words_wo_punct=''.join(no_punct)
    return words_wo_punct

df_train['punctuation']=df_train['message'].apply(lambda x: remove_punctuation(x))
df_train.head()
```

# 2.Tokenization

```python
def tokenize(text):
    split=re.split("\W+",text) 
    return split
df_train['tokenized']=df_train['punctuation'].apply(lambda x: tokenize(x.lower()))
df_train.head()
```

# 3.Stop words

```python
stopword = nltk.corpus.stopwords.words('english')
print(stopword[:11])
```

```python
def remove_stopwords(text):
    text=[word for word in text if word not in stopword]
    return text
df_train['stopwords'] = df_train['tokenized'].apply(lambda x: remove_stopwords(x))
df_train.head()
```

# 4.Lemmatisation

```python
#nltk.download('wordnet')
def train_lemma(words, lemmatizer):
    return [lemmatizer.lemmatize(word) for word in words]  

lemmatizer = WordNetLemmatizer()

df_train['lemma'] = df_train['stopwords'].apply(train_lemma, args=(lemmatizer, ))
```

```python
# view datase
df_train
```

# 5.Split the data 

```python
X = df_train['lemma']
y = df_train['sentiment']
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

# 6. vectorize the data 

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def clean_title(text):
   text = " ".join([word.lower() for word in text if word not in string.punctuation])
   title = re.split('\W+', text)
   text = [ps.stem(word) for word in title if word not in nltk.corpus.stopwords.words('english')]
   return text
```

```python
#create vector object
cv = CountVectorizer(analyzer=clean_title) 

# transfrom the values based on X_train
X_train_vect = cv.fit_transform(X_train).toarray()
```

```python
#transform the values X_test
X_test_vect = cv.transform(X_test).toarray()
```

```python
#convert y_train to array
y_train_vect = np.array(y_train)
```

# multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB 
```

```python
mn = MultinomialNB()
```

```python
mn.fit(X_train_vect, y_train_vect)
```

```python
y_pred = mn.predict(X_test_vect)
```

```python
print("The accuracy score of the multinomial Naive Bayes is:", accuracy_score(y_test, y_pred))
print("\n\nClassification Report:\n\n", classification_report(y_test, y_pred))
```
